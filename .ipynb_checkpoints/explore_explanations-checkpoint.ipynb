# SAE Neuron Explanation Explorer

Analysis of explanation quality for sparse autoencoder neurons trained on **Qwen 2.5-3B-Instruct**, explained by DeepSeek V3.2.

**Holdout structure per neuron:** 3 high-activation + 2 low-activation examples, deterministically sampled with seed 42069.

**Key rule:** Low-activation examples are a negative control. Prediction quality metrics (overlap, location, combined) are only meaningful for high-activation examples — we include low-activation only where explicitly relevant (Section 1.5 sanity check).

---
## Sections
1. Aggregations
2. Outliers
3. Questions

```python
import json, os, glob
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.dpi'] = 120

EXPL_DIR = 'explanations'
```

---
## 1. Aggregations

```python
rows = []
for path in sorted(glob.glob(os.path.join(EXPL_DIR, '*_unified_scores.json'))):
    with open(path) as f:
        data = json.load(f)
    nid = data['neuron_id']
    layer = int(nid.split('_')[0])
    for res in data.get('results', []):
        if not res.get('pairs'):
            continue
        for pair in res['pairs']:
            rows.append({
                'neuron_id': nid,
                'layer': layer,
                'example_idx': res['example_idx'],
                'normalized_activation': res['normalized_activation'],
                'activation_type': res.get('activation_type', 'unknown'),
                'prediction_rank': pair['prediction_rank'],
                'concept_score': pair['concept_score'],
                'meaning_score': pair['meaning_score'],
                'overlap_fraction': pair['overlap_fraction'],
                'location_score': pair['location_score'],
                'combined_score': pair['combined_score'],
            })

df = pd.DataFrame(rows)
df_high = df[df['activation_type'] == 'high']
df_low  = df[df['activation_type'] == 'low']

print(f"Total rows: {len(df)} across {df['neuron_id'].nunique()} neurons")
print(f"  High-activation rows : {len(df_high)}")
print(f"  Low-activation rows  : {len(df_low)}")
```

```python
# Per-neuron summary — HIGH examples only
neuron_best = df_high.groupby(['neuron_id', 'layer']).agg(
    best_combined  = ('combined_score',   'max'),
    mean_combined  = ('combined_score',   'mean'),
    best_meaning   = ('meaning_score',    'max'),
    mean_meaning   = ('meaning_score',    'mean'),
    best_overlap   = ('overlap_fraction', 'max'),
    mean_overlap   = ('overlap_fraction', 'mean'),
    best_concept   = ('concept_score',    'max'),
    mean_concept   = ('concept_score',    'mean'),
).reset_index()

print(f"{len(neuron_best)} neurons with scored high-activation examples")
neuron_best.describe().round(3)
```

```python
# Score distributions — HIGH examples only
score_cols = ['combined_score', 'meaning_score', 'overlap_fraction', 'location_score', 'concept_score']
fig, axes = plt.subplots(2, 3, figsize=(14, 8))

for ax, col in zip(axes.flat, score_cols):
    ax.hist(df_high[col], bins=50, alpha=0.7, edgecolor='black', linewidth=0.3)
    ax.axvline(df_high[col].mean(), color='red', linestyle='--',
               label=f'mean={df_high[col].mean():.3f}')
    ax.set_title(col)
    ax.legend(fontsize=8)

ax = axes.flat[5]
ax.hist(neuron_best['best_combined'], bins=50, alpha=0.7, edgecolor='black', linewidth=0.3, color='orange')
ax.axvline(neuron_best['best_combined'].mean(), color='red', linestyle='--',
           label=f'mean={neuron_best["best_combined"].mean():.3f}')
ax.set_title('best_combined per neuron')
ax.legend(fontsize=8)

fig.suptitle('Score Distributions — High-Activation Examples Only', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# Layer-wise comparison — HIGH examples only
layer_stats = neuron_best.groupby('layer')[['best_combined', 'mean_combined', 'best_meaning', 'best_overlap']].agg(
    ['mean', 'median', 'std', 'count']
)
layer_stats.columns = ['_'.join(c) for c in layer_stats.columns]
display(layer_stats.round(3))

fig, axes = plt.subplots(1, 3, figsize=(14, 4.5))
for ax, metric in zip(axes, ['best_combined', 'best_meaning', 'best_overlap']):
    layers = sorted(neuron_best['layer'].unique())
    data_by_layer = [neuron_best[neuron_best['layer'] == l][metric].values for l in layers]
    bp = ax.boxplot(data_by_layer, labels=layers, patch_artist=True)
    colors = plt.cm.viridis(np.linspace(0.2, 0.8, len(layers)))
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)
    ax.set_title(metric)
    ax.set_xlabel('Layer')

fig.suptitle('Score Distributions by Layer — High-Activation Examples Only', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# Rank analysis — HIGH examples only
# Does rank 1 consistently outperform rank 2 and 3?
rank_stats = df_high.groupby('prediction_rank')[['combined_score', 'meaning_score',
                                                  'overlap_fraction', 'concept_score']].mean()
print("Mean scores by prediction rank (high-activation examples):")
display(rank_stats.round(4))

best_rank = df_high.loc[df_high.groupby(['neuron_id', 'example_idx'])['combined_score'].idxmax()]
rank_win_pct = best_rank['prediction_rank'].value_counts(normalize=True).sort_index()
print("\nHow often each rank achieves the highest combined_score:")
print(rank_win_pct.round(3))
```

```python
# Sanity check: high vs low activation example scores
# We EXPECT high to score better on overlap/location (we're checking if the scorer is sensible).
# meaning_score may be similar since the full document context is shared.
comparison = df.groupby('activation_type')[['combined_score', 'meaning_score',
                                             'overlap_fraction', 'location_score']].mean()
print("Mean scores — high vs low activation examples:")
display(comparison.round(4))
```

---
## 2. Outliers

```python
# Load explanation previews
explanations = {}
for path in sorted(glob.glob(os.path.join(EXPL_DIR, '*_unified_scores.json'))):
    with open(path) as f:
        data = json.load(f)
    explanations[data['neuron_id']] = data.get('explanation_used', '')[:300]

neuron_best['explanation_preview'] = neuron_best['neuron_id'].map(explanations)

def print_neurons(title, subset):
    print(f"\n{'='*60}")
    print(title)
    print('='*60)
    for _, row in subset.iterrows():
        print(f"\n{row['neuron_id']} (layer {row['layer']}) | best={row['best_combined']:.3f} mean={row['mean_combined']:.3f}")
        print(f"  {row['explanation_preview']}...")

print_neurons("TOP 10 BEST (by best_combined on high examples)",
              neuron_best.nlargest(10, 'best_combined')[['neuron_id', 'layer', 'best_combined', 'mean_combined', 'explanation_preview']])

print_neurons("TOP 10 WORST (by best_combined on high examples)",
              neuron_best.nsmallest(10, 'best_combined')[['neuron_id', 'layer', 'best_combined', 'mean_combined', 'explanation_preview']])
```

```python
# Neurons where meaning_score is high but overlap_fraction is zero
# = model conceptually understands the neuron but consistently predicts the wrong text location
rank1_high = df_high[df_high['prediction_rank'] == 1]

high_meaning_no_overlap = rank1_high[
    (rank1_high['meaning_score'] > 0.75) & (rank1_high['overlap_fraction'] == 0.0)
]

print(f"Rank-1 predictions with meaning > 0.75 but overlap = 0: {len(high_meaning_no_overlap)}")
print(f"That's {100*len(high_meaning_no_overlap)/len(rank1_high):.1f}% of all rank-1 high-activation predictions")

# Scatter: meaning vs overlap for rank-1 high examples, colored by layer
fig, ax = plt.subplots(figsize=(8, 6))
sc = ax.scatter(rank1_high['overlap_fraction'], rank1_high['meaning_score'],
                c=rank1_high['layer'], cmap='viridis', alpha=0.4, s=15)
ax.axhline(0.75, color='red', linestyle='--', alpha=0.5, label='meaning=0.75')
ax.set_xlabel('overlap_fraction')
ax.set_ylabel('meaning_score')
ax.set_title('Rank-1 Predictions: Meaning vs Overlap (High-Activation Only)')
fig.colorbar(sc, ax=ax, label='Layer')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Neurons with unified score files but no pairs (simulation failed or empty predictions)
all_score_files = sorted(glob.glob(os.path.join(EXPL_DIR, '*_unified_scores.json')))
scored_neurons = set(df['neuron_id'].unique())
all_neurons_in_files = set()

for path in all_score_files:
    with open(path) as f:
        data = json.load(f)
    all_neurons_in_files.add(data['neuron_id'])

missing = all_neurons_in_files - scored_neurons
print(f"Total unified score files: {len(all_score_files)}")
print(f"Neurons with valid pairs:  {len(scored_neurons)}")
print(f"Neurons with no pairs:     {len(missing)}")

# Break down by layer
missing_by_layer = {}
for nid in missing:
    layer = int(nid.split('_')[0])
    missing_by_layer[layer] = missing_by_layer.get(layer, 0) + 1
print("\nMissing by layer:", dict(sorted(missing_by_layer.items())))
```

## 3. Questions

Things I noticed in the data that I want to understand better.

### Q1: Does explanation quality predict rank reliability?

Rank 1 wins ~53% of the time overall. For neurons where the explanation is clearly good (high `best_combined`), does rank 1 win more reliably? If yes, the rank signal is meaningful when the explanation is good. If not, the model is just guessing regardless.

```python
best_rank_per_example = df_high.loc[
    df_high.groupby(['neuron_id', 'example_idx'])['combined_score'].idxmax()
][['neuron_id', 'example_idx', 'prediction_rank']]

rank1_win_rate = best_rank_per_example.groupby('neuron_id').apply(
    lambda g: (g['prediction_rank'] == 1).mean()
).reset_index()
rank1_win_rate.columns = ['neuron_id', 'rank1_win_rate']

neuron_q1 = neuron_best[['neuron_id', 'best_combined', 'layer']].merge(rank1_win_rate, on='neuron_id')

fig, ax = plt.subplots(figsize=(8, 5))
sc = ax.scatter(neuron_q1['best_combined'], neuron_q1['rank1_win_rate'],
                c=neuron_q1['layer'], cmap='viridis', alpha=0.5, s=15)
ax.set_xlabel('best_combined (explanation quality)')
ax.set_ylabel('rank-1 win rate')
ax.set_title('Does explanation quality predict rank-1 reliability?')
fig.colorbar(sc, ax=ax, label='Layer')
plt.tight_layout()
plt.show()

corr = neuron_q1['best_combined'].corr(neuron_q1['rank1_win_rate'])
print(f"Correlation between best_combined and rank-1 win rate: {corr:.3f}")
```

### Q2: Is `concept_score` doing anything useful?

`concept_score` is uniformly low (~0.19–0.26) and barely separates rank 1 from rank 2/3. Is it correlated with any other signal, or is it effectively noise? If it's decorrelated from `overlap_fraction` and `meaning_score`, it might be measuring something genuinely independent — or it might just be poorly calibrated.

```python
score_cols = ['concept_score', 'meaning_score', 'overlap_fraction', 'location_score', 'combined_score']
corr = df_high[score_cols].corr()

fig, ax = plt.subplots(figsize=(7, 6))
im = ax.imshow(corr, cmap='RdBu_r', vmin=-1, vmax=1)
ax.set_xticks(range(len(score_cols)))
ax.set_yticks(range(len(score_cols)))
ax.set_xticklabels([c.replace('_', '\n') for c in score_cols], fontsize=9)
ax.set_yticklabels(score_cols, fontsize=9)
for i in range(len(score_cols)):
    for j in range(len(score_cols)):
        ax.text(j, i, f'{corr.iloc[i, j]:.2f}', ha='center', va='center', fontsize=9)
fig.colorbar(im, ax=ax, shrink=0.8)
ax.set_title('Score Correlation Matrix — High-Activation Examples')
plt.tight_layout()
plt.show()

print("concept_score variance:", df_high['concept_score'].var().round(5))
print("meaning_score variance:", df_high['meaning_score'].var().round(5))
```

### Q3: Why does layer 35 consistently underperform?

Layer 35 has the lowest mean `best_combined` and the highest variance. Is this because:
- Deep neurons encode more abstract/distributed features that are harder to name?
- There are fewer layer 35 neurons in the sample (selection bias)?
- The activation values themselves are more spread out, making the holdout harder?

Let's check activation magnitude distributions by layer.

```python
activation_by_layer = df_high.groupby('layer')['normalized_activation'].agg(['mean', 'std', 'median', 'count'])
print("High-activation example stats by layer:")
display(activation_by_layer.round(3))

fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))

# Activation magnitude by layer
layers = sorted(df_high['layer'].unique())
data_by_layer = [df_high[df_high['layer'] == l]['normalized_activation'].values for l in layers]
axes[0].boxplot(data_by_layer, labels=layers, patch_artist=True)
axes[0].set_title('Normalized Activation Magnitude by Layer (high examples)')
axes[0].set_xlabel('Layer')
axes[0].set_ylabel('normalized_activation')

# best_combined by layer
data_by_layer2 = [neuron_best[neuron_best['layer'] == l]['best_combined'].values for l in layers]
axes[1].boxplot(data_by_layer2, labels=layers, patch_artist=True)
axes[1].set_title('Best Combined Score by Layer')
axes[1].set_xlabel('Layer')
axes[1].set_ylabel('best_combined')

plt.tight_layout()
plt.show()
```

### Q4: Are there neurons where the explanation is confidently wrong?

A confidently wrong explanation = high `meaning_score` (model is confident about the concept) + zero `overlap_fraction` across ALL three predictions + low `combined_score`. These are the cases where the explainer produced a plausible-sounding but incorrect explanation — the simulator believes the explanation but keeps predicting the wrong part of the text.

```python
# Per-neuron: max meaning score and max overlap across all high-activation predictions
neuron_signals = df_high.groupby('neuron_id').agg(
    max_meaning  = ('meaning_score',    'max'),
    max_overlap  = ('overlap_fraction', 'max'),
    best_combined = ('combined_score',  'max'),
).reset_index()

# Confidently wrong: high meaning, zero overlap across all predictions
confidently_wrong = neuron_signals[
    (neuron_signals['max_meaning'] > 0.75) & (neuron_signals['max_overlap'] == 0.0)
].sort_values('max_meaning', ascending=False)

print(f"Neurons with max_meaning > 0.75 but zero overlap across all predictions: {len(confidently_wrong)}")
print()

confidently_wrong['explanation_preview'] = confidently_wrong['neuron_id'].map(explanations)
for _, row in confidently_wrong.head(5).iterrows():
    print(f"{row['neuron_id']} | max_meaning={row['max_meaning']:.3f} max_overlap=0.0 best_combined={row['best_combined']:.3f}")
    print(f"  {row['explanation_preview']}...")
    print()
```

### Q5: Do neurons that activate on specific/concrete concepts score higher than abstract ones?

Intuitively, a neuron for "the word 'right'" should be easier to simulate than one for "abstract forces." Can we see this in the data? The `concept` field in each prediction is a single word/phrase — if we cluster by concept specificity (short = specific, long = abstract), does score correlate?

```python
# Load concept labels from simulation files
concept_rows = []
for path in sorted(glob.glob(os.path.join(EXPL_DIR, '*_simulation.json'))):
    with open(path) as f:
        data = json.load(f)
    nid = data.get('neuron_id', '')
    for res in data.get('results', []):
        for pred in res.get('predictions', []):
            concept = pred.get('concept', '')
            if concept and pred.get('rank') == 1:
                concept_rows.append({
                    'neuron_id': nid,
                    'concept': concept,
                    'concept_word_count': len(concept.split()),
                })

concept_df = pd.DataFrame(concept_rows)
print(f"Loaded {len(concept_df)} rank-1 concept labels from {concept_df['neuron_id'].nunique()} neurons")

# Average concept word count per neuron → proxy for abstractness
neuron_concept_len = concept_df.groupby('neuron_id')['concept_word_count'].mean().reset_index()
neuron_concept_len.columns = ['neuron_id', 'mean_concept_words']

merged = neuron_best[['neuron_id', 'best_combined', 'layer']].merge(neuron_concept_len, on='neuron_id', how='inner')

fig, ax = plt.subplots(figsize=(8, 5))
sc = ax.scatter(merged['mean_concept_words'], merged['best_combined'],
                c=merged['layer'], cmap='viridis', alpha=0.5, s=15)
ax.set_xlabel('Mean concept label word count (rank-1 predictions)')
ax.set_ylabel('best_combined')
ax.set_title('Concept Specificity vs Explanation Quality')
fig.colorbar(sc, ax=ax, label='Layer')
plt.tight_layout()
plt.show()

corr = merged['mean_concept_words'].corr(merged['best_combined'])
print(f"Correlation (concept word count vs best_combined): {corr:.3f}")
```
